```python
fila_vendedores = []

while True:
  print("\n"+"="*40)
  print(" FILA DE VENDEDORES")
  print("="*40)
  print("1. Adicionar vendedor")
  print("2. Chamar próximo")
  print("3. Ver fila")
  print("4. Sair")

  opcao = input("\nOpção:")

  if opcao == "1":
    nome = input("Nome do vendedor:")
    fila_vendedores.append(nome)
    print(f" {nome}adicionado!")

  elif opcao =="2":
    if fila_vendedores:
      proximo = fila_vendedores.pop(0)
      print(f" Próximo: {proximo}")
    else:
      print(" Fila Vazia!")

  elif opcao =="3":
   if fila_vendedores:
     print("\nFila:")
     for i, v in enumerate(fila_vendedores, 1):
      print(f"{i}. {v}")
   else:
    print("Fila Vazia")

  elif opcao =="4":
    print("Encerrado...")
    break

else:
  print("Opção inválida!")



```
